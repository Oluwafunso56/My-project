```python
data.head()
```

```python
# Total customer by sex
plt.figure(figsize=(5,3))
sns.countplot(x=data['sex'])
```

```python
# Total customer by sex and smoker
plt.figure(figsize=(5, 3))
sns.countplot(x=data['smoker'], hue=data['sex'])
plt.show()
```

```python
# Total customer by smoker
plt.figure(figsize=(5,3))
sns.countplot(x=data['smoker'])
```

```python
# Total customer by day
plt.figure(figsize=(5,3))
sns.countplot(y=data['day'])
```

```python
from matplotlib.projections import axes
# Visualize the numerical columns - Total bill and Tips (Boxplot)
fig, axs = plt.subplots(1, 2, figsize=(15, 3))

sns.boxplot(data['total_bill'], ax=axs[0])
axs[0].set_title("Total bill")

sns.boxplot(data['tip'], ax=axs[1])
axs[0].set_title("Tip")



```

```python
data.head()
```

Bivariate Analysis- Based on two features(column) in the dataset


```python
# Total bill by sex
total_billsex = data.groupby('sex')['total_bill'].sum().reset_index()
sns.barplot(x='sex', y='total_bill', data=total_billsex)
```

```python
# is there relationship between total bill and tip
sns.scatterplot(x='total_bill', y='tip', data=data)
plt.figure(figsize=(5,3))
```

```python
# Total bill by sex and Total tips by sex
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Plot total bill by sex
axs[0].set_title('Total Bill by Sex')
total_bill_sex = data.groupby('sex')['total_bill'].sum().reset_index()
sns.barplot(x='sex', y='total_bill', data=total_bill_sex, ax=axs[0])

# Plot total tip by sex
axs[1].set_title('Total Tip by Sex')
total_tips_sex = data.groupby('sex')['tip'].sum().reset_index()
sns.barplot(x='sex', y='tip', data=total_tips_sex, ax=axs[1])
```

Multivariate Analysis- Based on two or more features (columns)

```python
# is there a relationship between total bill and tip by gender
plt.figure(figsize=(5,3))
```

```python
sns.scatterplot(x='total_bill', y='tip', data=data, hue='sex')
plt.figure(figsize=(5,3))
```

```python
sns.scatterplot(data=data, x='total_bill', y='tip', hue='sex')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Multivariate Analysis: Total Bill and Tip by Gender')

#show the plot
plt.show()
```

```python
sns.scatterplot(data=data, x='total_bill', y='tip', hue='day')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Multivariate Analysis: Total Bill and Tip by week')

#show the plot
plt.show()
```

```python
#size by gender
#size by sex
sizesex = data.groupby('sex')['size'].sum().reset_index()
sns.barplot(x='sex', y='size', data=sizesex)
```
